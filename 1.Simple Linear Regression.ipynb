```python
#Phase 1
#Library

import pandas as pd

    # pandas is a library funtion which has multiple functions in it, also it is globalized for the users.
    # After intializing pandas, all functions under pandas will be loaded in a particulr file memory.(while import pandas).
    # pandas can handle database related files like tables.
    # as means converting pandas(bigword) to pd (shortform) here.
```

```python
#2.Dataset

dataset=pd.read_csv("Salary_Data.csv")

    # pd is the class,read csv is the function called from the class pd.
    # The dataset file is passed within the bracket with double quotes.
    # File name sholud be same otherwise File not Found error will occur.
    # File not Found Error will occur while 1.misspelled the file name 2.If the dataset is not under the working folder.
    # csv means comma seperated value. 
```

```python
dataset

    # To view or display the content ie, columns and rows in the dataset file.
    # From the table input and output columns can be identified.
    # In this, the dataset using is preprocessed one so preprocessing step is not required.
```

```python
# 3.Fetching input & output columns

independent=dataset[["YearsExperience"]]

    # Assigning independent variable ie., input from the dataset.
    # Input column ie., YearsExperiene from the table is fetched from the dataset by giving the column name (without spelling mistake) inside the double square bracket within double quotes.
    # To fetch more than one column means use comma and give column name without spelling mistake.
```

```python
independent

    # To display the content under the  input column ie.,independent variable.
```

```python
dependent=dataset[["Salary"]]

    # To fetch/seperate the output column from the dataset by assinging the dependent variable. 
```

```python
dependent

    # To display the content under the output column from the dataset.
```

```python
# 4. Spliting training and testset

from sklearn.model_selection import train_test_split

    #sklearn supports machine leaning algorithms.
    
x_train,x_test,y_train,y_test=train_test_split(independent,dependent,test_size=0.30,random_state=0)

    # Importing only train nd test funtion from folder model selection under sklearn.
    # Here calling only the required function rather all like pandas,so the consumption of memory is less.
    # The positions while spliting 1.input training set(x_train) 2.input testset(x_test) 3.output training set(y_train) 4.output testset(y_test).
    # Underscore is used to make the single word simply insted of space.
    # For the function train_test_split the prameters are passed inside the bracket as independent & dependent with the testset size & random state.
    # Training & Test dataset can be splited in the ratio of 70:30,80:20 etc.
    # Training dataset is used to make the model learn.
    # Test set is used to evaluate the model.
```

```python
# 5. Learning phase
# SLR algorithmn

from sklearn.linear_model import LinearRegression

    # Simple Linear regression is a Machine learning algorithm provided by sklearn.
    # LinearRegression is called and imported from the funtion linear_model.
    
regressor=LinearRegression()

    # Creating variable name regressor (easy to follow),the functions under LinearRegression is assingned to the variable regressor.

regressor.fit(x_train,y_train)

    # Calling fit funtion through the access operator(.),the formula is substitute here for the model ie model is created using fit method.
    # For SLR, Y=wx+b
               # where w=weight
               #       x=input
               #       b=bias
```

```python
weight=regressor.coef_

    # The model value is checked by variable weight means the coefficient values of slope and bias will be saved in variable weight.
```

```python
bias=regressor.intercept_

    # To know the bias value using intercept_.

bias

    # To display the bias or intercept value.
```

```python
# 6. Evaluation

y_pred=regressor.predict(x_test)

    # The values of the model ie., weight and bias are saved in the variable regressor.
    # To evaluate the model input test dataset is used(x_test).
    # creating y_pred variable and evaluation is done on testset.
    
```

```python
from sklearn.metrics import r2_score

    # Getting evaluation result through r2_score function.
        
r_score=(y_test,y_pred)

    # r_score variable is created and input is giving ie., actual value (y_test) and predicted (y_pred).
    # r score value will be between 0-1, if the result is nearest to 1 then the model is good.
```

```python
r_score
```

```python
# 6. Saving the model

import pickle

    # Pickle is the function to save the model in python.
    
filename="finalized_model_linear.sav"

    # Assign file name as finalized _model_linear.sav within double quotes.
    # .sav is the pickle's extension filename.
```

```python
pickle.dump(regressor,open(filename,'wb'))

    # wb means write binary.
```

```python
# 7. Loading the model to check

loaded_model=pickle.load(open("finalized_model_linear.sav",'rb'))

    # loaded_model is the variable name created and loaded by the function pickle.load, rb is read binary.
    
result=loaded_model.predict([[15]])

    # To check the result enter the input value inside the double square bracket ie ., 15 here.
    
```
